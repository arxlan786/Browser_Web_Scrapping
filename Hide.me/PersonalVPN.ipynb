```python

class PersonalVPN_Account:
    
    
    def __init__(self):
    
        self.browser = webdriver.Chrome(executable_path='E:\My Task\hideme\chromedriver.exe')
        self.REMOTE_SERVER = "www.google.com"
        self.username = 'test'
        self.password = '201557404154'
        self.hideme_webpage = 'https://hide.me/en/'
        self.emailfake_webpage = 'https://emailfake.com/'
        
        
    def is_connected(self):
        try:
            host = socket.gethostbyname(self.REMOTE_SERVER)
            s = socket.create_connection((host, 80), 2)
            s.close()
            return True
        except:
            pass
        return False

                    
    def wait_for_keyPress(self):
        keypress.getch()
        
        
    def isBrowserOpen(self):
        try:
            self.browser.execute(Command.STATUS)
            return True
        except (socket.error , httplib.CannotSendRequest):
            return False
        
    
    def failed(self):
        print("Account Creation Failed!")
        #print("Press any key to Continue...")
        #self.wait_for_keyPress()
        if self.isBrowserOpen():
            self.browser.quit()
        sys.exit()
        print("Exit")
        
        
    def readNumber(self):
        try:
            with open('doc.txt', 'r') as file:
                self.acc_no = int(file.readline())+1
        except IOError as err:
            errno, strerror = err.args
            print("I/O error({0}): {1}".format(errno, strerror))
            print("Kindly check bot 'doc.txt'and 'PersonalVPN.py' are in same folder ", end='')
            time.sleep(0.3)
            print("Or You have to create 'doc.txt' text file and put text '1'")
            self.failed()
    
    def writeNumber(self):
        try:
            with open('doc.txt','w') as file:
                file.write(str(self.acc_no))
        except IOError as err:
            errno,sterror = err.args
            print("I/O error ({0}):{1}".format(errno,strerror))
            print("Kindly check bot 'doc.txt'and 'PersonalVPN.py' are in same folder ", end='')
            time.sleep(0.3)
            print("Or You have to create 'doc.txt' text file and put text '1'")
            self.failed()
        
    

    
    
    def makeAccount(self):
        temp = 0
        internet_retry = 5
        for i in tnrange(internet_retry , desc = 'Checking Internet Connection'):
            time.sleep(0.01)
            if self.is_connected():
                temp +=1
        if temp != internet_retry:
            print("Check your internet connection...")
            self.failed()
            
        else:
            self.readNumber()

            # ------------------------------------------------------ #  
            #                                                        #
            #              Opening HideMe WebPage                    #
            #                                                        #
            # ------------------------------------------------------ #  
    
            print("Open HideMe WebPage for Email Registeration")
            self.browser.get(self.hideme_webpage)
            hideme_tab = self.browser.window_handles[0]
            
            # Finding Yellow Download Button
            download_btn = self.browser.find_element_by_xpath("//div[@class='Promo-holder']/a[@class='c-ctabtn']")
            # Click the Download Button
            download_btn.send_keys(Keys.ENTER)
            
            # Finding Register Button in this page
            register_btn = self.browser.find_element_by_xpath("//div[@class='o-pricingtable__content u-pt10']/a[@class='c-btn c-btn__light c-btn--md u-fullwidth u-uppercase js-tracked']")
            # Click the Register Button
            register_btn.send_keys(Keys.ENTER)

            # ------------------------------------------------------ #  
            #                                                        #
            #              Opening Fake Email WebPage                #
            #                                                        #
            # ------------------------------------------------------ #  

            print("Open Fake Email Webpage")
            self.browser.execute_script("window.open('about:blank', 'emailfake_tab');")
            self.browser.switch_to.window('emailfake_tab')
            self.browser.get(self.emailfake_webpage)
            
            # Finding Green Copy Button on this page
            copy_btn = self.browser.find_element_by_xpath("//div[@class='e7m coserch']/button[@class='e7m btn btn-success waves-effect waves-light waves-raised']")
            # Click the Copy Button
            copy_btn.send_keys(Keys.ENTER)
            
            # ------------------------------------------------------ #  
            #                                                        #
            #      Copy Fake Email and Paste it to HideMe_Tab        #         
            #                                                        #
            # ------------------------------------------------------ #  
            
            print("Copying Fake Email and Paste it to Hide.me Resgisteration Tab")
            self.browser.switch_to.window(hideme_tab)
            
            # Finding Signup Fields Fields
            signup_input = self.browser.find_element_by_xpath("//div[@class='form-group']/input[@class='form-control js-registration_email']")
            signup_input.send_keys(Keys.CONTROL + 'v')
            signup_Reg_btn = self.browser.find_element_by_xpath("//form[@class='page-member__form']/button[@class='submit-btn js-registration_btn']")
            signup_Reg_btn.send_keys(Keys.ENTER)
            
            # ------------------------------------------------------ #  
            #                                                        #
            # Now Move towards Email Fake Tab for Email Confirmation #         
            #                                                        #
            # ------------------------------------------------------ #  
            
            self.browser.switch_to.window('emailfake_tab')
            
            try:
                print("Waiting for Email Confirmation...")
                Activate_btn = WebDriverWait(self.browser , 100).until(
                    EC.presence_of_element_located(
                        (By.XPATH,"//a[text()='Activate my account']"))
                )
                
            except:
                print('Error Recieving Confirmation Email from hide.me')
                self.failed()
            
            Activate_btn.send_keys(Keys.ENTER)
            print("Email Activated/Confirmed")
            
            # ------------------------------------------------------ #  
            #                                                        #
            #          Now Create Email in HideMe SignupTab          #         
            #                                                        #
            # ------------------------------------------------------ #  
            
            signUp_hideme_tab = self.browser.window_handles[2]
            self.browser.switch_to.window(signUp_hideme_tab)
            try:
                print("Signup in hide.me ")
                signUp_hideme_inp = WebDriverWait(self.browser,100).until(
                    EC.presence_of_element_located(
                        (By.XPATH,"//div[@class='form-group']/input[@class='form-control js-activation-username']"))
                )
                signUp_hideme_inp.send_keys(str(self.username)+str(self.acc_no))
                time.sleep(2)
            
            except:
                print("Fatal Error! while SignUp in Hide.me Membership Tab")
                self.failed()
            
            while (True):
                try:
                    signUp_hideme_inp_err = WebDriverWait(self.browser,4).until(
                        EC.presence_of_element_located(
                            (By.XPATH,"//div[@class='form-group activation_field_notvalid']/span[@class='activation_message activation_error']"))
                    )
                    print("Email Already Exist..." , end='')
                    time.sleep(0.01)
                    print("Don't be panic this will generate auto new email :)")
                    self.writeNumber()
                    self.readNumber()
                    signUp_hideme_inp = self.browser.find_element_by_xpath("//div[@class='form-group activation_field_notvalid']/input[@class='form-control js-activation-username not-empty']")
                    signUp_hideme_inp.send_keys(Keys.CONTROL+'a')
                    signUp_hideme_inp.send_keys(Keys.BACK_SPACE)
                    signUp_hideme_inp.send_keys(str(self.username)+str(self.acc_no))
                    time.sleep(1)
                except:
                    print("Email Generated... :) ")
                    break
                   
            signUp_hideme_pass = self.browser.find_element_by_xpath("//div[@class='form-group']/input[@class='form-control js-activation-password']")
            signUp_hideme_pass.send_keys(self.password)
            signUp_hideme_cpass = self.browser.find_element_by_xpath("//div[@class='form-group']/input[@class='form-control js-activation-password2']")
            signUp_hideme_cpass.send_keys(self.password)
            
            # Click Create Account Button
            signUp_hideme_btn = self.browser.find_element_by_xpath("//div[@class='page-activate__fields']/button[@class='submit-btn']")
            signUp_hideme_btn.send_keys(Keys.ENTER)
            print("Account Created... :)")
            self.writeNumber()
            try:
                _ = WebDriverWait(self.browser , 100).until(
                    EC.presence_of_element_located((By.XPATH,"//a[@class='o-headertop__select js-headertoponclick']"))
                )
                
            except:
                print("Error in Creating Hide.me VPN Account")
                self.failed()
            
            self.browser.quit()        
            print('New Personal VPN account is created successfully.')
            print(f'Username: {str(self.username) + str(self.acc_no)}')
            print(f'Password: {str(self.password)}')
            
                 
if __name__ == "__main__":
    
    new = PersonalVPN_Account()
    new.makeAccount()
    
        
        
        
```
